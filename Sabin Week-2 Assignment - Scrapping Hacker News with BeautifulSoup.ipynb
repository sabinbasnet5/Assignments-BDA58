```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.ycombinator.com/'
html = requests.get(url)
print(html.text)
```

```python
soup = BeautifulSoup(html.content, 'html.parser')
soup.title.string #Getting title name 
```

```python
# Find all the article headlines and links
headlines = soup.find_all('tr', class_='athing')
print(headlines)
```

```python
# Extract headlines and links for each article
for article in articles:
  # Find the title element
  title_element = article.find('span', class_='titleline')
  
  # Extract headline text
  headline = title_element.find('a').text.strip()
  
  # Extract link URL
  link = title_element.find('a')['href']
  
  # Print the extracted headline and link
  print(f"Headline: {headline}")
  print(f"Link: {link}")
  print("---")
```

```python
# Prepare data for CSV
data = []
for article in articles:
    title_element = article.find('span', class_='titleline')
    headline = title_element.find('a').text.strip()
    link = title_element.find('a')['href']
    data.append([headline, link])
 # Print the extracted headline and link
data
```
